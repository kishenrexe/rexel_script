```python
def combine_columns(row):
    if row ['kenmerkwaarde'] > '' and row['kenmerkwaarde max'] > 0:
        return row['kenmerkwaarde'] + ' ('+ row['kenmerkwaarde max'] + ')'
    elif row['kenmerkwaarde'] > '':
        return row['kenmerkwaarde']
    elif row['kenmerkwaarde max'] > '':
        return '(' + row['kenmerkwaarde max'] + ')'
    else: 
        return ' '
```

```python
df['combined'] = df.apply(combine_columns, axis=1)

df
```
