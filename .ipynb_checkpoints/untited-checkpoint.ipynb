```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_excel('EC011141.xlsx')




```

```python
data1["kenmerkwaarde"] = data1["kenmerkwaarde"].astype(str) + '  ' + data1["kenmerkwaarde max"].astype(str)
data1["waarde"] = data1["waarde"] + ' ' + data1["eenheid"].astype(str)
```

```python
data1["waarde"] = data1["waarde"] + ' (' + data1["eenheid"].astype(str) + ')'

data1
```

```python
data2 = data1.pivot(index='code', columns='waarde', values=['kenmerkwaarde','kenmerkwaarde max'])

data2
```

```python
data2.dropna(how='all')
```

```python
data2.replace(np.nan,' ')
```

```python
result = pd.concat([data1, data2], axis=1).reindex(data1.index)

result
```

```python
df3=pd.merge(data1,data2, left_on='code', right_on='code')

df3
```

```python
df3.to_excel("output.xlsx") 
```

```python
df3 =df3.drop(columns=['waarde', 'kenmerkwaarde', 'kenmerkwaarde max', 'eenheid', 'etim value', 'Unnamed: 0'])

df3
```

```python
df2 = df3.drop_duplicates(keep='first')

df2
```

```python
df2.to_excel("output.xlsx") 
```

```python
df2= df2.dropna(how="all", axis=1)

df2
```

```python
df2 =df2.drop(columns=df2.columns[0], axis=1,  inplace=True)

df2
```

```python
del df2[df2.columns[0]]
```
